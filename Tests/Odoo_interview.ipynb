```python
def look_and_say(integer):
    int_str = str(integer)
    cur_char = int_str[0]
    cur_count = 1
    str_result = ''
    for new_char in int_str[1:]:
        if new_char == cur_char:
            cur_count = cur_count + 1
        else:
            str_result = str_result + str(cur_count) + cur_char
            cur_count = 1
            cur_char = new_char
    str_result = str_result + str(cur_count) + cur_char
    return int(str_result)

look_and_say(1135527778)
```

```python
def common(a, b, c):
    sum = 0
    i = 0
    for i in range(len(a)):
        if a[i] in b and a[i] in c:
            sum += a[i]
            b.pop(b.index(a[i]))
            c.pop(c.index(a[i]))
    return sum

# len(a)*(len(b) + len(c)) ~ square?
# important: integer 'types', numbers of repetition
# unimportant: order, non-repeated in other lists

def common_alt(a, b, c):
    # dict index = integer, dict value = number of repetitions
    sum = 0
    dicts = []
    for list in (a, b, c):
        temp_dict = {}
        for integer in list:
            temp_dict[integer] = temp_dict.setdefault(integer, 0) + 1
        dicts.append(temp_dict)
    #print(dicts)
    for key in dicts[0]: sum += min(dicts[0].get(key, 0), dicts[1].get(key, 0), dicts[2].get(key, 0))*key
    return sum

# len(a) + len(b) + len(c)
# len(keys_a)*( """ get """ x3) ~ linear

common_alt([1,2,2,3],[5,3,2,2],[7,3,2,2])
```

```python
# Folder structure parsing: determine the size of every folder
# Look for an 'ls' -> determine where we are (perhaps first) -> look at the output -> sum up the files
# -> Record it as ~path with size

class Dir():
    def __init__(self, parent_dir, name, path):
        self.parent_dir = parent_dir
        self.name = name
        self.path = path
        self.local_size = 0
        self.size = 'UNDEFINED'
        self.subdirs = {}

    def total_size(self):
        size = self.local_size
        for value in self.subdirs.values(): size += value.total_size()
        return size

    def print_info(self):
        print("Dir: " + self.name + " at path " + str(self.path) + " with subdirs " + str(self.subdirs.keys()) + " with size of " + str(self.total_size()))

input_file = open("Test.txt", 'r')
#print(type(input_file))
lines = []
line = input_file.readline()
while line != '':
    lines.append(line[:-1].split(' '))
    line = input_file.readline()
#print(lines)
#input()
# Now I have a list of line elements
cur_path = ['/']
root_dir = Dir('NONE', '/', cur_path)
cur_dir = root_dir
i = 1 # ignoring the $ cd /
sum = 0
while i < len(lines):
    #print("Iteration: " + str(i) + " Line: " + str(lines[i]))
    #cur_dir.print_info()
    if lines[i][0] == '$': # command
        if lines[i][1] == 'cd':
            if lines[i][2] != '..': # lines[i][2] is the new dir name here
                cur_path.append(lines[i][2])
                cur_dir = cur_dir.subdirs[lines[i][2]]
            else:
                if cur_dir.size == 'UNDEFINED':
                    cur_dir.size = cur_dir.total_size() # Leaving a dir = we explored everything in it = we can count it right now, with the flag proving it only happens once
                    if cur_dir.size <= 100000: sum += cur_dir.size
                cur_path.pop()
                cur_dir = cur_dir.parent_dir
        elif lines[i][1] == 'ls': # iterate over ls output until we reach the next command
            j = i + 1
            while lines[j][0] != '$':
                if lines[j][0] == 'dir':
                    # Check if this subdir was already made (for repeated $ ls)
                    if lines[j][1] not in cur_dir.subdirs: cur_dir.subdirs[lines[j][1]] = Dir(cur_dir, lines[j][1], cur_path)
                else: cur_dir.local_size += int(lines[j][0])
                if j == len(lines) -1: break
                else: j += 1
            i = j - 1
    i += 1
# Return to root
while cur_dir != 'NONE':
    if cur_dir.size == 'UNDEFINED':
        cur_dir.size = cur_dir.total_size() # Leaving a dir = we explored everything in it = we can count it right now, with the flag proving it only happens once
        if cur_dir.size <= 100000: sum += cur_dir.size
    cur_path.pop()
    cur_dir = cur_dir.parent_dir
#root_dir.print_info()
#for dir in root_dir.subdirs.values(): dir.print_info()
print("Total sum is:", sum)
```
